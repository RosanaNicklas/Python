```python
import twint
from textblob import TextBlob
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import nest_asyncio
nest_asyncio.apply()
```

```python

# Transformar los resultados en pandas df
def twint_to_pandas(columns):
    return twint.output.panda.Tweets_df[columns]
```

```python
# Configuracion de Twint:
c = twint.Config()
c.Search = "Pablo Iglesias"
c.Limit = 10
c.Pandas = True
```

```python
twint.run.Search(c)
df = twint_to_pandas(["date", "username",
 "tweet", "hashtags", "nlikes"])
```

# An치lisis de Sentimiento - Sentiment Analysis

```python

list_sentiment = []
for tweet in df["tweet"]:
 # print(tweet)
 analysis = TextBlob(tweet)
 # print(analysis.sentiment)

 if analysis.sentiment[0] > 0:
    list_sentiment.append(1.0)
   #print('Positive')
 elif analysis.sentiment[0] < 0:
    list_sentiment.append(2.0)
 # print(tweet)
 #print('Negative')
 else:
    list_sentiment.append(0.0)
    #print("Neutro")
    #print('\n')
print(list_sentiment)
df["Sentiment"] = list_sentiment
```

```python
df
```

```python
sns.distplot(df["Sentiment"])
sns.set(rc={"figure.figsize": (10,6)})
```

```python
df_new = df[["Sentiment"]].value_counts()
df_new
```

```python
df_new = df.groupby(['Sentiment']).count()
df_new
```
